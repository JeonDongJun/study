```python
# list를 dic형식으로 예제
list_example = ['a', 'b', 'c', 'd', 'a', 'b', 'c', 'a', 'b', 'a']

remove_overlap = set(list_example)

dic = {}
 
for list_name in remove_overlap:
    dic[list_name] = 0

for list_count in list_example:
    dic[list_count] = dic[list_count] + 1
    
print(dic)
```

```python
#konlpy예제
from konlpy.tag import Kkma
from konlpy.utils import pprint

kkma = Kkma()
print(kkma.morphs(u'공부를 하면할수록 모르는게 많다는 것을 알게 됩니다.'))

print(kkma.nouns(u'대학에서 DB, 통계학, 이산수학 등을 배웠지만...'))

print(kkma.pos(u'다 까먹어버렸네요?ㅋㅋ'))

print(kkma.sentences(u'그래도 계속 공부합니다. 재밌으니까!'))
```
